A diferencia de la regresi√≥n lineal, en este modelo encontraremos la mejor funci√≥n log√≠stica para describir la relaci√≥n entre variable dependiente e independiente.

En otras palabras, el modelo de regresi√≥n log√≠stica predice P(Y=1) en funci√≥n de "X".



    

**¬øPorqu√© existe un modelo diferente para este tipo de modelo?**

Pues b√°sicamente porque la regresi√≥n lineal no da una buena l√≠nea de ajuste para los problemas que tienen solo dos niveles en la variable dependiente, y consecuentemente da menos precisi√≥n durante la predicci√≥n porque no cubre los conjuntos de datos, ya que es de naturaleza lineal.

Es as√≠ que para lograr un mejor ajuste de los datos categ√≥ricos, se requiere una curva que s√≥lo es posible con la regresi√≥n log√≠stica, ya que utiliza una funci√≥n sigmoidea para hacer predicciones.

**La funci√≥n log√≠stica**

La funci√≥n log√≠stica tambi√©n conocida como Sigmoida generalmente se denota por œÉ(x) o sig(x) y esta dada por:


$$œÉ(x) = \frac{1}{1+e^{-x}}$$



Esta es la funci√≥n matem√°tica que tiene la 'curva en forma de S', y su valor siempre se encuentra entre 0 y 1, por lo que se la implementa para resolver problemas categ√≥ricos que tienen dos valores posibles.

Adicionalmente, establece un valor para punto de corte, que en su mayor√≠a se establece en 0,5, por lo que si se supera el resultado previsto de la curva log√≠stica, p > 0.5, la nueva observaci√≥n se clasificar√° como y = 1, de lo contrario se clasificar√° como y = 0.

Tras c√°maras, lo que sucede es que la Regresi√≥n log√≠stica funciona como la lineal, en el sentido que transforma los resultados obtenidos por esta aplicando la funci√≥n sigmoidea para acotar los valores entre 0 y 1 y as√≠ garantizar que las probabilidades est√©n siempre entre 0 y 1.

As√≠, cuando "X" alcanza valores muy altos y positivos, el valor de $e^{-x}$ es aproximadamente 0. Si reemplazamos esto en la f√≥rmula precedente veremos entonces que, el valor de la funci√≥n sigmoide es 1. Por el contrario, cuando "X" alcanza valores muy altos pero negativos, el valor de $e^{-x}$ tiende a infinito por lo que el valor de la funci√≥n sigmoide ser√° de 0.

Ejemplo:
    
    Por ejemplo, si x = 10, entonces f(x) = 1 / (1 + e^-10) = 1 / (1 + 0.000045) = 0.999955, lo que significa que la funci√≥n sigmoide nos da un resultado cercano a 1.

    Pero si x = -10, entonces f(x) = 1 / (1 + e^10) = 1 / (1 + 22026) = 0.000045, lo que significa que la funci√≥n sigmoide nos da un resultado cercano a 0.

En resumen, cuando x es grande y positivo, la funci√≥n sigmoide se acerca a 1, y cuando x es grande y negativo, la funci√≥n sigmoide se acerca a 0.

**Conceptos previos: Odds y logaritmo de los odds**

En estad√≠stica, los odds son la relaci√≥n entre la probabilidad de que algo suceda y la probabilidad de que no suceda, por lo que la f√≥rmula se la describe como:

$$odds = \frac{(P) de. que. algo. suceda}{(P) de. que. algo. no. suceda}$$


 $$ odds=\frac{p}{1-p} $$

Por ejemplo, la probabilidad de que una observaci√≥n pertenezca a la clase y = 1 es $\frac{p}{1-p}$, por lo que cuando las probabilidades est√°n entre 0 y 1, los odds est√°n en contra de que la observaci√≥n pertenezca a y = 1. En cambio, cuando las probabilidades son mayores que 1, las odds est√°n a favor de la observaci√≥n pertenezca a y = 1.

Pongamos un ejemplo num√©rico, si p=0.2, entonces las odds son 0.2/0.8 = 0.25, lo que significa que las odds est√°n en contra de que la observaci√≥n pertenezca a la clase "y=1". Pero si p=0.8, entonces las odds son 0.8/0.2 = 4, lo que significa que las odds est√°n a favor de que la observaci√≥n pertenezca a la clase "y=1".

_Nota No. 1: Los odds son un s√≠mil a ratios de probabilidad y no est√°n acotados entre 0 y 1._

Este concepto es un poco m√°s f√°cil de comprender si lo pensamos bajo la √≥ptica de apuestas. Por ejemplo: si estamos en un casino y queremos jugar a los dados, podr√≠amos apostar a que saldr√≠a un n√∫mero, en este caso escojamos el n√∫mero 6.

La probabilidad de que ganemos es 1/6. Entonces, los odds a favor de que ganemos son (1/6) / (5/6) = 1/5 o 1:5, como esto equivale a 0.2, y no es mayor a 1,  las probabilidades est√°n claramente en contra de que ganemos dado que las probabilidades de que perdamos son (5/6) / (1/6) = 5:1, lo que equivale a 5.

El rango de valores que pueden tomar los Odds es de [0,‚àû].

Aplicando la transformaci√≥n logit, podemos convertir el rango de probabilidades limitado de [0,1] a [‚àí‚àû,+‚àû].

Ojo aqu√≠: los odds est√°n acotados entre 0 e infinito, ya que el odds es la relaci√≥n entre la probabilidad de un evento y la probabilidad complementaria, que siempre est√°n entre 0 y 1.

Pero log odds est√°n acotados entre menos infinito y m√°s infinito, ya que el logaritmo de cualquier n√∫mero positivo es siempre mayor que menos infinito y menor que m√°s infinito. Incluso cuando la probabilidad de un evento es 0 o 1, lo que resultar√≠a en un odds de 0 o infinito, el log odds se puede calcular y tiene un valor finito de menos infinito o m√°s infinito, respectivamente.

Esta es una de las ventajas del uso de log odds en la estad√≠stica, ya que permite trabajar con valores que pueden ser negativos, lo que no es posible con odds simples que s√≥lo tienen valores positivos.

Otra ventaja del uso de log odds es que se pueden interpretar directamente como incrementos en las probabilidades logar√≠tmicas, lo que facilita la interpretaci√≥n de los resultados.

Por su parte el logaritmo de los odds, no es m√°s que lo que su nombre sugiere: el logaritmo natural de la funci√≥n previamente descrita. A este √∫ltimo se lo conoce tambi√©n como la funci√≥n log√≠stica y explicada a m√°s detalle en la secci√≥n del modelo matem√°tico.

 $$ odds=\frac{p}{1-p} $$

f(ln(p / (1-p))) = p

La √∫ltima expresi√≥n nos dice que la funci√≥n log√≠stica convierte los log odds en una probabilidad. Esto es importante porque la probabilidad es una medida m√°s f√°cil de interpretar que los log odds.

Por ejemplo, si el odds de que un equipo gane un partido es 3 (es decir, la probabilidad de que el equipo gane es 3/4), entonces el log odds es ln(3) = 1.0986. Si aplicamos la funci√≥n log√≠stica a este log odds, obtenemos:

f(1.0986) = 0.7502

Esto significa que la probabilidad de que el equipo gane es del 75%.

**Definici√≥n matem√°tica del modelo**

Tomando en cuenta la funci√≥n logit, podemos definir a la ecuaci√≥n log√≠stica como:

$$logit(p) = log(odds)=logit(Y)=\ln (\frac{p}{1-p})=\hat \beta_0+\hat \beta_{1}X_1+.... \hat \beta_{n}X_n+\underbrace{\epsilon}_{error}$$


Ahora, si reemplazamos el valor de X con el de la funci√≥n lineal obtenemos:

$$logit(Y)=\frac{1}{1+e^-{(\hat \beta_0+\hat \beta_{1}X_1)}}$$


Resolviendo para el denominador obtenemos la probabilidad de que la variable cualitativa "Y" tome el valor k (el nivel de referencia, codificado como 1), dado que el predictor "X" tiene el valor x.

$$logit(Y)=\frac{e^{(\hat \beta_0+\hat \beta_{1}X_1)}}{1+e^{(\hat \beta_0+\hat \beta_{1}X_1)}}$$

Donde Œ≤0  es la probabilidad esperada de pertenecer a la clase 1 cuando todos los predictores son cero. Por su parte, el resto de betas indican el cambio promedio del logaritmo de odds al incrementar en una unidad la variable predictora "X", manteniendo constantes el resto de variables.

Existe una relaci√≥n con esta funci√≥n y el logaritmo de los Odds, que nos permite relacionar de mejor manera los estimadores:



$$ ln(\frac{{p(Y=k|X=x)}}{{(1-p(Y=k|X=x)}})=\hat \beta_0+\hat \beta_{1}X_1+.... +\underbrace{\epsilon}_{error}$$


Recordemos que en la regresi√≥n lineal simple, model√°bamos el valor de la variable dependiente "Y" en funci√≥n del valor del predictor "X". Sin embargo, en la Regresi√≥n Log√≠stica, se modela la _probabilidad_ de que nuestra variable de respuesta "Y" pertenezca al nivel de referencia 1 en funci√≥n del valor de las variables explicativas, mediante el uso de Log of Odds expresado matem√°ticamente en la anterior funci√≥n.

**Interpretaci√≥n**

**Interpretaci√≥n de los Coeficientes en la Regresi√≥n Log√≠stica**

En la regresi√≥n log√≠stica, los coeficientes beta (Œ≤) representan el cambio en los log odds de la variable dependiente (Y) asociado con un cambio unitario en la variable independiente (X). La interpretaci√≥n de estos coeficientes depende del tipo de variable independiente que se est√© considerando: num√©rica o categ√≥rica.

**Variables Independientes Num√©ricas**

Para una variable independiente **num√©rica**, los coeficientes beta indican el cambio en los log odds de la variable dependiente por cada incremento unitario en la variable independiente.

Por ejemplo, si tenemos un modelo de regresi√≥n log√≠stica que predice la probabilidad de que una persona tenga diabetes en funci√≥n de su edad, y el coeficiente beta para la variable **edad** es 0.05, esto significa que por cada a√±o adicional de edad, los log odds de tener diabetes aumentan en 0.05. Dicho de otra forma, el odds de tener diabetes se multiplica por \(e^{0.05}\) (aproximadamente 1.05) por cada a√±o adicional.

**Variables Independientes Categ√≥ricas**

Para una variable independiente **categ√≥rica**, los coeficientes beta se interpretan como el cambio en los log odds entre dos niveles de la variable categ√≥rica.

- **Variables Categ√≥ricas con Dos Niveles**  
  Si la variable categ√≥rica tiene dos niveles, como el g√©nero (masculino o femenino), el coeficiente beta representa la diferencia en los log odds entre los dos g√©neros.

  Por ejemplo, si tenemos un modelo de regresi√≥n log√≠stica que predice la probabilidad de que una persona compre un producto en funci√≥n de su **g√©nero**, y el coeficiente beta para el g√©nero es -0.8, esto indica que los log odds de compra son 0.8 unidades menores para las mujeres que para los hombres (asumiendo que el g√©nero femenino est√° codificado como 1).

- **Variables Categ√≥ricas con M√°s de Dos Niveles**  
  Cuando la variable categ√≥rica tiene m√°s de dos niveles, el coeficiente beta compara los log odds de cada nivel con respecto al nivel de referencia, que es el nivel omitido en la modelaci√≥n.

  Por ejemplo, si tenemos un modelo de regresi√≥n log√≠stica que predice la probabilidad de que una persona compre un producto seg√∫n su **nivel de ingresos** (bajo, medio o alto), y el coeficiente beta para el nivel de ingresos *medio* es 0.6, esto indica que los log odds de compra son 0.6 unidades mayores para las personas con ingresos medios en comparaci√≥n con las personas con ingresos bajos, manteniendo el nivel de ingresos *alto* como el nivel de referencia.



Al igual que en la regresi√≥n lineal, el modelo de regresi√≥n log√≠stica encuentra los mejores valores de los estimadores para ajustarse al conjunto de datos, la diferencia principal es que no lo hace por el m√©todo de MCO. Como el resultado
de la ecuaci√≥n lineal log√≠stica es log(p/(1-p)), no podemos calcular su valor por el m√©todo de regresi√≥n tradicional porque no conocemos el valor de p. El √∫nico resultado que conocemos es la clase de y = 0 o 1, por ello tenemos que emplear otro m√©todo de estimaci√≥n: el M√©todo de M√°xima Verosimulitud-MLE


**El m√©todo de M√°xima Versonilitud-MLE**

En este m√©todo de estimaci√≥n, se utiliza una funci√≥n de probabilidad que mide qu√© tan bien se ajusta un conjunto de par√°metros a una muestra de datos. Los valores de los par√°metros que maximizan la funci√≥n de verosimilitud son las estimaciones de m√°xima verosimilitud. En otras palabras, el objetivo es hacer inferencias sobre la poblaci√≥n que es m√°s probable que haya generado el conjunto de datos de entrenamiento.

En el caso de la regresi√≥n log√≠stica, queremos encontrar las estimaciones de los par√°metros Œ≤0 y Œ≤1.

Otra forma para ajustar un modelo de Regresi√≥n Log√≠stica es empleando descenso de gradiente, que es materia de la siguiente semana donde obtendremos resultados de estos modelos en el √°mbito del machine learning.

**Ajuste del modelo**

Al igual que en el modelo de regresi√≥n lineal, en el modelo de Regresi√≥n Log√≠stica analizaremos los p-values para evaluar si el modelo es mejor que un modelo sin predictores, solo con Œ≤0. Para ello nos basaremos en el likelihood ratio, que es el que Python utiliza por defecto.

El likelihood ratio usa la diferencia entre la probabilidad de obtener los valores observados con el modelo log√≠stico y las probabilidades de hacerlo con un modelo sin relaci√≥n entre las variables. Para ello, calcula la significancia de la diferencia de residuos entre el modelo con predictores y el modelo nulo. El estad√≠stico tiene una distribuci√≥n chi-cuadrado con grados de libertad equivalentes a la diferencia de grados de libertad de los dos modelos comparados. Si se compara respecto al modelo nulo, los grados de libertad equivalen al n√∫mero de predictores del modelo generado.

## Supuestos del modelo de Regresi√≥n Log√≠stica

Similar a la regresi√≥n lineal, la regresi√≥n log√≠stica tiene ciertos supuestos. Pero la mayor√≠a de ellos tambi√©n se pueden relajados con la transformaci√≥n logar√≠tmica de la mano izquierda.


### 1.Variable dependiente categ√≥rica binaria

Como se mencion√≥ al principio, la regresi√≥n log√≠stica es para problemas de clasificaci√≥n, por lo que necesitamos asegurarnos de que la variable objetivo sea binaria.

#### Formas de verificaci√≥n

- La forma m√°s sencilla de verificar el supuesto, es comprobando que en nuestra base de datos la variable "Y" sea categ√≥rica, o de tipo "object".

#### Posibles soluciones

Para dar soluci√≥n a este supuesto, lo mejor es recodificar nuestra variable dependiente a los valores de que algo suceda (1) y a los valores de que algo no suceda (0). En otras palabras, el factor de nivel 1 de la variable dependiente deber√≠a representar el resultado deseado.

### 2.Relaci√≥n lineal entre el logaritmo natural de odds y la variable continua

El modelo log√≠stico requiere que las variables independientes est√°n linealmente relacionadas con las probabilidades logar√≠tmicas, patrones en forma de U son una clara violaci√≥n de esta condici√≥n.

Importante: La Regresi√≥n Log√≠stica no requiere que la variable independiente presente una distribuci√≥n normal.

### 3.No colinealidad perfecta entre las covariables  

Aunque encontrar multicolinealidad perfecta es bastante extra√±o, lo que generalmente implica es que existe duplicidad de variables explicativas en el modelo. Tambi√©n puede ser el caso que existe una segunda variable derivada de la primera m√°s una constante.

El problema de la multicolinearidad perfecta es que implica estimadores indeterminados y una varianza infinita de los estimadores.

$$ var \hat{\beta} = inf $$

Lo cierto es que es m√°s com√∫n en la pr√°ctica encontrarnos con situaciones donde exista multicolinearidad no perfecta que provoque errores elevados y de acuerdo con (Gujarati, 2007, p.329) estad√≠sticos peque√±os que llevan a no rechazar la hp√≥tesis nula, lo que incrementar√≠a la probabilidad de cometer errores de tipo II.


En s√≠ el problema de violar este supuesto es que obtiene estimadores menos precisos y muy sensibles a cambios en los datos.

#### Formas de verificaci√≥n

Existe una herramienta llamada Factor de inflaci√≥n de la varianza-VIF la cual devuelve un valor conforme al nivel de asociaci√≥n de las variables predictoras ( mide la correlaci√≥n y la fuerza de la correlaci√≥n ). Los rangos a tomar en cuenta del VIF son los siguientes:

$$ \textrm{VIF}_i = \frac{1}{1-R_i^2}$$

- Si el VIF es 1, no existe entre las variables analizadas

- Si el VIF es >1 y <5, la correlaci√≥n es moderada entre entre las variables analizadas, pero no constituye a menudo una transgresi√≥n grave para requerir atenci√≥n

- Si el VIF es >5, la correlaci√≥n es potencialmente severa entre entre entre las variables analizadasuna por lo que las estimaciones de los coeficientes y los valores p en el resultado de la regresi√≥n probablemente no sean confiables

#### Posibles soluciones

- Verificar que no exista duplicidad entre variables

- Retirar uno de los pares de variables que presente el mayor VIF e ir trabajando uno a uno hasta que el VIF se estabilice

### 4.Grandes tama√±os de muestra

Se espera que el modelo de regresi√≥n log√≠stica sea constru√≠do con tama√±os de muestra bastante grandes, aunque dependiendo de los resultados del modelo, este supuesto puede ser relajado.

_Nota No. 2: Primero, la regresi√≥n log√≠stica no requiere que los t√©rminos de error se distribuyan normalmente. Segundo, no requiere homocedasticidad, por lo que consume menos tiempo en la verificaci√≥n de los supuestos._

## üîç **Enfoque de Ciencia de Datos: Regresi√≥n Log√≠stica**

Como hab√≠amos mencionado previamente, el enfoque de **ciencia de datos** se centra en **elegir los mejores par√°metros** y configurar adecuadamente el modelo para que la **estimaci√≥n de la variable dependiente** sea la mejor posible. En este sentido, la **regresi√≥n log√≠stica** es una herramienta clave para abordar problemas de **clasificaci√≥n**, donde el objetivo es predecir la probabilidad de que ocurra un evento determinado (por ejemplo, **si un cliente comprar√° o no** un producto).

La **optimizaci√≥n de los par√°metros** se realiza ajustando los **coeficientes** del modelo para minimizar una funci√≥n de costo, com√∫nmente basada en la **entrop√≠a cruzada**. En el enfoque de ciencia de datos, no solo nos interesa la precisi√≥n del modelo, sino tambi√©n la **interpretabilidad** y la capacidad de explicar c√≥mo cada caracter√≠stica afecta la probabilidad de un determinado resultado. Esto implica evaluar cuidadosamente m√©tricas como la **curva ROC**, **AUC**, y la **matriz de confusi√≥n**, para asegurarnos de que el modelo no solo tenga un buen rendimiento, sino que sea adecuado para la **toma de decisiones** dentro del contexto empresarial.


## M√©tricas de evaluaci√≥n

Por su parte, si el problema es de clasificaci√≥n, las m√©tricas m√°s comunes son la matriz de confusi√≥n y cocientes derivados de esta. Entre los m√°s populares encontramos a:
- Accuracy
- Presici√≥n
- Exhaustividad o "recall"
- F1 score

#### Accuracy

La precisi√≥n en los problemas de clasificaci√≥n es el n√∫mero de predicciones correctas realizadas por el modelo sobre todos los tipos de predicciones realizadas.

$$ Accuracy = \frac{TP+TN}{TP+TN+FP+FN} $$

En el Numerador, est√°n nuestras predicciones correctas (Verdaderos positivos y Verdaderos negativos) y en el denominador, todas las predicciones hechas por el algoritmo (tanto las correctas como las incorrectas).

_¬øCu√°ndo usar el accuracy?_
El accuracy es es una buena medida cuando las clases de variables de destino en los datos est√©n casi equilibradas.

#### Precisi√≥n o "precision"

La precisi√≥n es una medida que nos da luces sobre la calidad del modelo ya que eval√∫a, el n√∫mero de elementos identificados correctamente como positivos sobre el total de positivos.

$$ Precision = \frac{TP}{TP+FP} $$

#### Exhaustividad o "recall"

Es una medida que nos dice qu√© cantidad el modelo de machine learning es capaz de identificar. Responde a la pregunta ¬øqu√© porcentaje de la categor√≠a objetivo somos capaces de identificar?

$$ Recall = \frac{TP}{TP+FN} $$

#### F1 score

Es una puntuaci√≥n √∫nica que represente tanto la Precisi√≥n (P) como la exhaustividad o "recall".

$$ F1 = \frac{2*Precision*Recall}{Precision+Recall} = \frac{2*TP}{2*TP+FP+FN} $$

# **2.M√°quinas de Soporte Vectorial (SVM)**

## ¬øQu√© es SVM?

Las **M√°quinas de Soporte Vectorial (SVM)** son un poderoso algoritmo de **aprendizaje supervisado** utilizado principalmente para **clasificaci√≥n** y, en menor medida, para **regresi√≥n**. La idea principal detr√°s de SVM es encontrar un **hiperplano** en un espacio de **dimensi√≥n n** que pueda separar claramente los puntos de diferentes clases con el **margen m√°ximo** posible.

### Enfoque Conceptual

En un problema de **clasificaci√≥n binaria**, la SVM trata de encontrar el **hiperplano √≥ptimo** que mejor separe las dos clases. Un hiperplano es una l√≠nea en el caso de datos en dos dimensiones o un plano en datos tridimensionales, y as√≠ sucesivamente. La **separaci√≥n √≥ptima** significa maximizar el **margen** (la distancia m√≠nima entre el hiperplano y los puntos de las dos clases).

![SVM Hiperplano](https://raw.githubusercontent.com/andreaescobar111/udla-aanlitica-pred-2024/main/S3_Clasificacion/1.Presesion/imagenes/svm.jpeg)

### F√≥rmula Matem√°tica

El hiperplano se define mediante la siguiente ecuaci√≥n lineal:

$$
w \cdot x + b = 0
$$

Donde:
- **$w$** es un vector de pesos.
- **$x$** es el vector de caracter√≠sticas.
- **$b$** es el t√©rmino de sesgo o intersecci√≥n.

Para que los puntos de clase positiva y negativa est√©n correctamente clasificados y el margen sea m√°ximo, la SVM intenta resolver el siguiente problema de **optimizaci√≥n**:

$$
\min \frac{1}{2} \| w \|^2
$$

Sujeto a la condici√≥n de que para todos los puntos:

- **$y_i (w \cdot x_i + b) \geq 1$**, donde **$y_i$** es la etiqueta de clase ($+1$ o $-1$) y **$x_i$** son los puntos de datos.

Este enfoque busca **maximizar el margen**, que equivale a **minimizar la norma** de $w$.

### Hiperplano √ìptimo y Vectores de Soporte

El **hiperplano √≥ptimo** es aquel que tiene el mayor margen posible entre las dos clases. Los puntos de datos m√°s cercanos al hiperplano se denominan **vectores de soporte**. Estos vectores son cruciales porque determinan la posici√≥n del hiperplano.

- **Margen**: La distancia entre los vectores de soporte y el hiperplano.
- **Vectores de Soporte**: Puntos cr√≠ticos que est√°n m√°s cerca del l√≠mite de decisi√≥n.

### Hiperplano √ìptimo y Separaci√≥n

![Hiperplano SVM](https://raw.githubusercontent.com/andreaescobar111/udla-aanlitica-pred-2024/main/S3_Clasificacion/1.Presesion/imagenes/svm2.png)

En el gr√°fico, se puede ver c√≥mo el **hiperplano** (la l√≠nea s√≥lida) divide las dos clases con el **m√°ximo margen** posible. Los puntos en los bordes del margen son los **vectores de soporte**, que determinan la posici√≥n de este hiperplano.

## SVM con Datos No Lineales

Cuando los datos no son **linealmente separables**, SVM utiliza un enfoque conocido como el **truco del kernel** para mapear los datos a un espacio de mayor dimensi√≥n, donde sea m√°s f√°cil separarlos. Un **kernel** es una funci√≥n que convierte el problema no lineal en uno **lineal** en un espacio de dimensi√≥n mayor.

### Tipos de Kernels Comunes

1. **Kernel Lineal**: Para datos que se pueden separar linealmente.
2. **Kernel Gaussiano (RBF)**: Es un kernel que mapea los puntos de datos a un espacio dimensional infinito para hacer los datos linealmente separables.
3. **Kernel Polinomial**: Mapea los datos a un espacio con interacciones polin√≥micas.

La funci√≥n kernel m√°s utilizada es el **Radial Basis Function (RBF)**, que permite separar incluso conjuntos de datos complejos:

$$
K(x_i, x_j) = \exp\left(-\frac{\| x_i - x_j \|^2}{2\sigma^2}\right)
$$

## Par√°metros Importantes

1. **C (Regularizaci√≥n)**: Controla la penalizaci√≥n de los puntos mal clasificados. Un valor alto de **C** reduce el margen para clasificar correctamente todos los puntos, lo que puede llevar a **sobreajuste**.
2. **Gamma (en RBF)**: Controla la influencia de un solo punto de entrenamiento. Un valor bajo de **gamma** significa que cada punto tiene una gran influencia, mientras que un valor alto implica que la influencia est√° m√°s concentrada.

### Visualizaci√≥n del Kernel

En el siguiente gr√°fico, se muestra c√≥mo el uso de un **kernel RBF** permite que los datos, que inicialmente no eran separables linealmente, se puedan clasificar correctamente:

![Kernel RBF](https://raw.githubusercontent.com/andreaescobar111/udla-aanlitica-pred-2024/main/S3_Clasificacion/1.Presesion/imagenes/svm3.png)

## Ventajas y Desventajas de SVM

### Ventajas
- **Eficaz en espacios de alta dimensi√≥n**.
- Funciona bien cuando el n√∫mero de dimensiones es mayor que el n√∫mero de muestras.
- Utiliza un subconjunto de los puntos de entrenamiento (vectores de soporte), por lo que es **eficiente en t√©rminos de memoria**.

### Desventajas
- No funciona bien con conjuntos de datos **grandes**, ya que el tiempo de entrenamiento es **cuadr√°tico** en el n√∫mero de muestras.
- No es ideal cuando hay mucho **ruido** en los datos, ya que intenta encontrar un margen lo suficientemente amplio, lo que puede resultar en clasificaciones err√≥neas.

## Conclusi√≥n

Las **M√°quinas de Soporte Vectorial (SVM)** son herramientas potentes y vers√°tiles para **problemas de clasificaci√≥n** y se utilizan ampliamente en la **ciencia de datos** debido a su capacidad para crear clasificaciones robustas mediante la maximizaci√≥n del **margen** y el uso de **kernels** para datos no lineales. Con el ajuste adecuado de par√°metros como **C** y **gamma**, las SVM pueden lograr un rendimiento excelente, incluso en conjuntos de datos complejos y de alta dimensionalidad.


# **3. K-Nearest Neighbors (KNN)**

## ¬øQu√© es KNN?

El algoritmo de **K-Nearest Neighbors (KNN)** es uno de los m√©todos m√°s simples y efectivos de **aprendizaje supervisado** usado tanto para **clasificaci√≥n** como para **regresi√≥n**. KNN clasifica un punto de datos bas√°ndose en la **proximidad** con respecto a sus vecinos m√°s cercanos. Es un algoritmo **no param√©trico**, lo que significa que no hace ninguna suposici√≥n sobre la **distribuci√≥n de los datos**.

El concepto central de KNN es que "los objetos similares se encuentran cerca entre s√≠". El valor de **$K$** determina cu√°ntos "vecinos" se utilizar√°n para la predicci√≥n.

![Gr√°fico de KNN](https://raw.githubusercontent.com/andreaescobar111/udla-aanlitica-pred-2024/main/S3_Clasificacion/1.Presesion/imagenes/knn.png)

## Funcionamiento del Algoritmo

### 1. Selecci√≥n del Valor de K
El primer paso en el uso de KNN es seleccionar el valor de **$K$**, que es el n√∫mero de vecinos m√°s cercanos que se tendr√°n en cuenta para clasificar un nuevo punto. Un valor **peque√±o de $K$** puede hacer que el modelo sea **sensible al ruido** (sobreajuste), mientras que un valor **muy alto** puede llevar a un **subajuste** del modelo.

- **$K = 1$**: Clasificaci√≥n basada solo en el vecino m√°s cercano.
- **$K > 1$**: La clase se decide seg√∫n la mayor√≠a de los vecinos m√°s cercanos.

### 2. C√°lculo de la Distancia
Para determinar los **vecinos m√°s cercanos**, el algoritmo KNN utiliza una **m√©trica de distancia**. La m√°s com√∫n es la **distancia Euclidiana**, definida por:

$$
d(x_i, x_j) = \sqrt{\sum_{k=1}^{n} (x_{i,k} - x_{j,k})^2}
$$

Donde:
- **$x_i$** y **$x_j$** son los puntos de datos que se est√°n comparando.
- **$n$** es el n√∫mero de caracter√≠sticas.

### 3. Clasificaci√≥n o Regresi√≥n
- Para la **clasificaci√≥n**, KNN asigna al nuevo punto la **clase m√°s frecuente** entre los $K$ vecinos m√°s cercanos.
- Para la **regresi√≥n**, la predicci√≥n se calcula tomando la **media** (o alguna medida similar) de los valores de los vecinos m√°s cercanos.

### Visualizaci√≥n del Algoritmo

![KNN Clasificaci√≥n](https://raw.githubusercontent.com/andreaescobar111/udla-aanlitica-pred-2024/main/S3_Clasificacion/1.Presesion/imagenes/knn2.png)

En la imagen anterior, la clasificaci√≥n del punto nuevo (rojo) depende de los **vecinos m√°s cercanos** (en azul y verde). Seg√∫n la mayor√≠a de los vecinos m√°s cercanos, el punto ser√° asignado a la clase **verde** o **azul**.

## Ejemplo Conceptual

Supongamos que estamos clasificando frutas bas√°ndonos en caracter√≠sticas como el **peso** y el **color**. Si una fruta nueva tiene caracter√≠sticas similares a un grupo de manzanas, KNN la clasificar√° como una **manzana** si la mayor√≠a de los vecinos cercanos pertenecen a esa clase.

### Ventajas y Desventajas

#### Ventajas
1. **Simplicidad**: KNN es muy f√°cil de entender e implementar.
2. **Versatilidad**: Puede ser usado tanto para **clasificaci√≥n** como para **regresi√≥n**.
3. **No Param√©trico**: No hace suposiciones sobre la distribuci√≥n de los datos.

#### Desventajas
1. **Costoso en T√©rminos de Computaci√≥n**: KNN tiene que calcular la distancia a cada punto de entrenamiento, lo cual se vuelve **ineficiente** para conjuntos de datos grandes.
2. **Sensibilidad a la Escala de los Datos**: Caracter√≠sticas con escalas mayores pueden tener m√°s peso al calcular la distancia. Es recomendable **escalar** los datos antes de usar KNN.
3. **Elecci√≥n del Valor de K**: Un valor **incorrecto de $K$** puede llevar a **sobreajuste** o **subajuste**.

### Selecci√≥n del Valor de K

La elecci√≥n del **mejor valor de $K$** es crucial para el rendimiento del modelo. En la pr√°ctica, el valor de $K$ suele determinarse mediante **validaci√≥n cruzada**, buscando un equilibrio entre **sesgo** y **varianza**.

- **Valores Bajos de K** (ej. $K = 1, 3$): Pueden ser **demasiado sensibles** al ruido, lo que conduce a **sobreajuste**.
- **Valores Altos de K** (ej. $K > 20$): Pueden **promediar** demasiada informaci√≥n, lo que hace que el modelo sea demasiado **simple** y pierda detalles importantes.

## Aplicaciones Comunes de KNN

1. **Clasificaci√≥n de Im√°genes**: Clasificar im√°genes basadas en caracter√≠sticas extra√≠das.
2. **Detecci√≥n de Anomal√≠as**: Detectar puntos de datos que se encuentran lejos de sus vecinos, indicando posibles **anomal√≠as**.
3. **Recomendaci√≥n**: Sistemas de recomendaci√≥n, donde los productos similares son sugeridos bas√°ndose en el comportamiento del usuario.

## Conclusi√≥n

**K-Nearest Neighbors (KNN)** es un algoritmo extremadamente **vers√°til** y **f√°cil de implementar**, ideal para problemas donde se necesita una **soluci√≥n r√°pida** y la interpretaci√≥n es importante. Sin embargo, se debe tener cuidado al elegir el valor de **$K$** y al **escalar los datos** adecuadamente, especialmente cuando el n√∫mero de caracter√≠sticas o muestras es alto, para asegurar que el rendimiento del modelo sea √≥ptimo.

#### Comparar el Resultado con el OneHotEncoder

El LabelEncoder es una t√©cnica de codificaci√≥n que asigna un n√∫mero entero √∫nico a cada valor categ√≥rico en la variable. A diferencia de la codificaci√≥n "dummy" o "one-hot", el LabelEncoder codifica una variable categ√≥rica en una sola columna, lo que puede hacer que el modelo trate la variable como una variable num√©rica.

El problema de utilizar el LabelEncoder es que introduce un orden num√©rico en los valores categ√≥ricos, lo cual no siempre es apropiado en todas las situaciones. En algunos casos, el orden de los valores categ√≥ricos no tiene un significado l√≥gico o no se puede establecer un orden claro. En este caso, el LabelEncoder puede confundir al modelo y producir resultados inexactos.

En resumen, se recomienda tener precauci√≥n al utilizar el LabelEncoder en la regresi√≥n log√≠stica, y preferir t√©cnicas de codificaci√≥n como la codificaci√≥n "dummy" o "one-hot" si se desea modelar adecuadamente una variable categ√≥rica con m√∫ltiples niveles.
