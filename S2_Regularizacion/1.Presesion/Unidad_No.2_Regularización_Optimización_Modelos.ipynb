## üîç Contenidos

### üîó Regularizaci√≥n y Sobreajuste
La **regularizaci√≥n** es una estrategia utilizada para mejorar la **generalizaci√≥n** de los modelos de regresi√≥n. En modelos complejos o con muchas variables, es com√∫n que el modelo aprenda demasiado bien los datos de entrenamiento, perdiendo la capacidad de **predecir correctamente** en nuevos conjuntos de datos. Para solucionar esto, utilizamos t√©cnicas de regularizaci√≥n que imponen **penalizaciones** a los coeficientes para controlar la complejidad del modelo.

### üßÆ T√©cnicas de Regularizaci√≥n
- **Ridge Regression (Regresi√≥n Ridge)**: Tambi√©n conocida como **regresi√≥n de cresta**, Ridge a√±ade una **penalizaci√≥n L2** a la suma de los cuadrados de los coeficientes, lo cual ayuda a reducir el valor de los coeficientes muy grandes y a evitar el sobreajuste. Ridge es √∫til cuando hay muchas **variables correlacionadas** entre s√≠.
  
- **Lasso Regression (Least Absolute Shrinkage and Selection Operator)**: Lasso, en cambio, utiliza una **penalizaci√≥n L1** que tiende a reducir algunos coeficientes a **cero**, lo que efectivamente selecciona un subconjunto de variables importantes, proporcionando un modelo m√°s **sencillo y f√°cil de interpretar**.

### ‚öñÔ∏è Balanceando Sesgo y Varianza
Otro concepto clave que veremos es c√≥mo la **regularizaci√≥n** permite balancear el **sesgo** y la **varianza** en el modelo. Un modelo con **alto sesgo** no captura la complejidad de los datos, mientras que un modelo con **alta varianza** se ajusta demasiado a los datos espec√≠ficos de entrenamiento. Al aplicar t√©cnicas de **Ridge** y **Lasso**, aprenderemos a encontrar el punto √≥ptimo que minimice ambos.

## ü§î ¬øPor Qu√© Aprender Sobre Regularizaci√≥n?
Las t√©cnicas de regularizaci√≥n son esenciales en el **an√°lisis predictivo**, especialmente cuando trabajamos con conjuntos de datos que tienen un n√∫mero elevado de **variables explicativas** o cuando hay correlaciones entre ellas. Ridge y Lasso permiten ajustar el modelo para encontrar la mejor **compensaci√≥n entre ajuste y simplicidad**, asegurando una mayor capacidad de **predicci√≥n**.

# üìä Regularizaci√≥n Ridge y Lasso en Regresi√≥n Lineal

La **regularizaci√≥n** es una t√©cnica utilizada para evitar el **sobreajuste** de los modelos de regresi√≥n. Cuando un modelo se ajusta demasiado bien a los datos de entrenamiento, puede perder la capacidad de generalizar para nuevos datos. **Ridge** y **Lasso** son dos tipos de regularizaci√≥n que nos ayudan a mejorar la capacidad predictiva y la estabilidad de los modelos lineales al a√±adir una **penalizaci√≥n** a los coeficientes.

## üîó Regularizaci√≥n Ridge (L2)

La **regresi√≥n Ridge**, tambi√©n conocida como **regularizaci√≥n L2**, a√±ade una penalizaci√≥n proporcional al **cuadrado de los coeficientes** a la funci√≥n de costo del modelo de regresi√≥n lineal. El objetivo es minimizar la suma de los errores cuadrados mientras se penalizan los coeficientes para que no crezcan de forma desproporcionada.

La funci√≥n objetivo de Ridge es:

$$
J(\beta) = \sum_{i=1}^{n} (y_i - \hat{y_i})^2 + \lambda \sum_{j=1}^{p} \beta_j^2
$$

Donde:
- $J(\beta)$: Funci√≥n de costo (Error Cuadr√°tico).
- $(y_i - \hat{y_i})^2$: Error entre el valor real $y_i$ y el valor predicho $\hat{y_i}$.
- $\lambda$: Par√°metro de regularizaci√≥n (tambi√©n llamado **alpha** en Python). Controla la fuerza de la penalizaci√≥n. Si $\lambda = 0$, se convierte en una regresi√≥n lineal sin regularizaci√≥n.
- $\sum_{j=1}^{p} \beta_j^2$: Penalizaci√≥n L2 que suma el cuadrado de los coeficientes.

**Interpretaci√≥n**:
- La **regularizaci√≥n Ridge** tiende a **reducir** los valores de los coeficientes sin hacerlos cero. Esto ayuda especialmente cuando se tiene **multicolinealidad** (cuando las variables predictivas est√°n altamente correlacionadas).

### üìà Gr√°fico de Coeficientes Ridge
En la visualizaci√≥n de los coeficientes de Ridge, a medida que **$\lambda$** aumenta, los coeficientes se reducen. Sin embargo, **nunca llegan a ser exactamente cero**. Esto es √∫til cuando se tiene un gran n√∫mero de caracter√≠sticas y se desea conservarlas todas, pero evitando que algunas dominen demasiado al modelo.

## üìâ Regularizaci√≥n Lasso (L1)

La **regresi√≥n Lasso** (Least Absolute Shrinkage and Selection Operator), o **regularizaci√≥n L1**, a√±ade una penalizaci√≥n proporcional al **valor absoluto de los coeficientes**. Esto no solo reduce el valor de los coeficientes, sino que puede llevar algunos de ellos a **cero**, lo que hace que el modelo sea **m√°s sencillo e interpretable**.

La funci√≥n objetivo de Lasso es:

$$
J(\beta) = \sum_{i=1}^{n} (y_i - \hat{y_i})^2 + \lambda \sum_{j=1}^{p} |\beta_j|
$$

Donde:
- $\sum_{j=1}^{p} |\beta_j|$: Penalizaci√≥n L1 que suma el valor absoluto de los coeficientes.

**Interpretaci√≥n**:
- La **regularizaci√≥n Lasso** tiende a llevar algunos coeficientes a **cero**, lo cual hace que Lasso sea √∫til para la **selecci√≥n de variables**. Si algunos coeficientes se convierten en cero, significa que esas variables se eliminan del modelo, lo que resulta en un modelo m√°s interpretable.

### üìà Gr√°fico de Coeficientes Lasso
En el caso de Lasso, a medida que aumentamos el valor de **$\lambda$**, algunos coeficientes se reducen a **cero**. Esto permite hacer una **selecci√≥n autom√°tica** de las variables m√°s importantes, y es especialmente √∫til cuando se sospecha que muchas variables son irrelevantes.

## üìä Comparaci√≥n Visual entre Ridge y Lasso

- **Ridge**: Reduce todos los coeficientes, pero no elimina ninguno. Es preferible cuando **todas las caracter√≠sticas son relevantes** y est√°n **correlacionadas** entre s√≠.
- **Lasso**: Puede eliminar algunos coeficientes llev√°ndolos a **cero**, lo que equivale a una **selecci√≥n autom√°tica** de caracter√≠sticas. Es √∫til cuando se necesita un modelo m√°s **simple** y cuando algunas variables no aportan mucho al modelo.

### üîç ¬øCu√°ndo Usar Ridge o Lasso?
- **Ridge**: Cuando todas las caracter√≠sticas son importantes y tienen cierta correlaci√≥n entre s√≠, ya que Ridge mantiene todas las variables en el modelo, pero con coeficientes m√°s peque√±os.
- **Lasso**: Cuando sospechas que **solo algunas variables** son importantes. Lasso reduce a cero los coeficientes irrelevantes, lo cual ayuda a la selecci√≥n de variables.

### üìã Ejemplo Pr√°ctico

Supongamos que tenemos un **conjunto de datos de viviendas** y queremos predecir el precio de una casa usando caracter√≠sticas como el **n√∫mero de habitaciones**, **tama√±o**, **n√∫mero de ba√±os**, etc. Utilizando **Ridge**, podemos asegurarnos de que ninguna caracter√≠stica tenga un peso desproporcionado, mientras que con **Lasso** podr√≠amos eliminar autom√°ticamente caracter√≠sticas como la **edad** de la casa si se demuestra que no tiene impacto significativo en el precio.
